```python
import pandas as pd
import numpy as np

import datetime as dt

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
logs = pd.read_json('../data/logins.json')
```

```python
logs.info()
```

```python
logs = logs.set_index('login_time')
```

```python
logs['count'] = 1
```

```python
logs.sort_index().head(10)
```

```python
logs15 = logs.resample('15t').sum()
logs15.head(10)
```

# Total logins

```python
logs15['count'].plot(title='Total count')
plt.ylabel('number of logins')
plt.xlabel('time')

plt.show()
```

# Logins by month 

```python
logs15.loc['1970-01','count'].plot(title='Count by month')
logs15.loc['1970-02','count'].plot()
logs15.loc['1970-03','count'].plot()
logs15.loc['1970-04','count'].plot()

plt.ylabel('number of logins')
plt.xlabel('time')

plt.savefig('../figures/monthcountplot.png')
```

```python
# max logins by month
print('January:', logs15.loc['1970-01','count'].max())
print('February:', logs15.loc['1970-02','count'].max())
print('March:', logs15.loc['1970-03','count'].max())
print('April:', logs15.loc['1970-04','count'].max())
```

```python
logs15.loc['1970-01','count'].plot(title='January logins')

plt.ylabel('number of logins')
plt.xlabel('time')
plt.xticks(np.arange(31))

plt.show()
```

```python
logs15.loc['1970-02','count'].plot(title='February logins')

plt.ylabel('number of logins')
plt.xlabel('time')

plt.show()
```

```python
logs15.loc['1970-03','count'].plot(title='March logins')

plt.ylabel('number of logins')
plt.xlabel('time')

plt.show()
```

```python
logs15.loc['1970-04','count'].plot(title='April logins')

plt.ylabel('number of logins')
plt.xlabel('time')

plt.show()
```

```python
logs15.loc['1970-02-26':'1970-03-02', 'count'].plot()

plt.ylabel('number of logins')
plt.xlabel('time')

plt.show()
```

```python
series = logs15.reset_index()['login_time']
```

```python
series.dt.dayofweek[:20]
```

```python
logs15['day_of_week'] = series.dt.dayofweek.values
```

```python
logs15.sample(10)
```

```python
groupby_dow = logs15.groupby('day_of_week').mean()
```

```python
avg_count = groupby_dow.sort_values('count', ascending=False)
avg_count
```

```python
logs15.loc['1970-02-26':'1970-03-02', ['count', 'day_of_week']].sort_values('count',ascending=False).head(10)
```

```python
# people most dependably login on Friday according to the average count; however, the largest numbers of logins
# have occured on Saturdays. 
```

# Logins by time of day

```python
logs15['day'] = logs15.reset_index()['login_time'].dt.day.values
```

```python
logs15.sample(10)
```

```python
January = logs15['1970-01'].groupby('day')['count'].sum()
February = logs15['1970-02'].groupby('day')['count'].sum()
March = logs15['1970-03'].groupby('day')['count'].sum()
April = logs15['1970-04'].groupby('day')['count'].sum()
```

```python
January.plot()
February.plot()
March.plot()
April.plot()

plt.legend(['Jan', 'Feb', 'Mar', 'Apr'])

plt.show()
```

```python
# the day of the month likely unrelated to number of logins
```

```python
logs15['hour'] = logs15.reset_index()['login_time'].dt.hour.values
logs15['minute'] = logs15.reset_index()['login_time'].dt.minute.values
```

```python
logs15.sample(10)
```

```python
logs15.groupby('hour')['count'].sum().sort_values(ascending=False)
```

```python
logs15.groupby('hour')['count'].sum().plot()

plt.savefig('../figures/hourofdayplot.png')
```

```python
# number of logins is highest from 10 pm to 3 am and again 10:30 am to 1 pm
```
